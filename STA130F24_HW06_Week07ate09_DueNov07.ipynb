The Simple Linear Regression model predicts a continuous outcome variable Y based on a single predictor variable X. It assumes a linear relationship between X and Y with an equation that can be written as: 
$$ Y = \beta_0 + \beta_1x + \epsilon $$
The predictor variable X is the independent variable that we use to predict the outcome variable. The outcome variable Y is the dependent variable, which we are trying to predict. 𝛽0 is the intercept, which represents the starting point of the line on the Y axis (the value of Y when X=0). 𝛽1 is the slope, which is the coefficient represents the change in Y for one unit increase in X, and it also determines the direction and steepness of the line. 𝜖 is the error term, accounts for the variability in Y that cannot be explained by X. In theory, 𝜖 is assumed to follow a normal distribution with a mean of zero and some standard deviation 𝜎.

Each observed Y value is generated by adding this error term to the linear prediction of 𝛽0 + 𝛽1*X. This makes Y a random variable that is normally distributed for any given.

````python
import numpy as np
import plotly.graph_objects as go
from scipy.stats import norm

# Set parameters
n = 100  # Number of data points
beta_0 = 5  # Intercept
beta_1 = 2  # Slope
sigma = 1  # Standard deviation of the error term

# Generate predictor variable X
np.random.seed(0)  # for reproducibility
x = np.random.uniform(0, 10, size=n)  # 100 random values between 0 and 10

# Generate the error term epsilon ~ N(0, sigma^2)
epsilon = np.random.normal(0, sigma, size=n)

# Generate outcome variable Y based on the Simple Linear Model equation
Y = beta_0 + beta_1 * x + epsilon

# Create plot
fig = go.Figure()

# Scatter plot of simulated data points
fig.add_trace(go.Scatter(x=x, y=Y, mode='markers', name='Simulated Data'))

# Line for the theoretical regression (without error)
Y_true = beta_0 + beta_1 * x
fig.add_trace(go.Scatter(x=x, y=Y_true, mode='lines', name='True Line (Y = β0 + β1*x)'))

# Plot layout
fig.update_layout(title="Simple Linear Regression Model (Theoretical) with scipy.stats",
                  xaxis_title="Predictor (x)",
                  yaxis_title="Outcome (Y)",
                  showlegend=True)

fig.show(renderer="png")
````

Intercept 𝛽0 and Slope 𝛽1 are set to 5 and 2, respectively, meaning our theoretical regression line is Y = 5 + 2x. By setting n=100, we  generate 100 random values of x from a uniform distribution between 0 and 10. ϵ is drawn from a normal distribution with mean 0 and standard deviation of 1, simulating the random noise. Outcome Variable Y is calculated using the simple linear regression equation, which combines the linear term 𝛽0 + 𝛽1X and the error term 𝜖. By plot the simulated data points and the true regression line (without error term) to visualize how the error term causes the observed values to deviate from the line.

````python
import scipy.stats as stats
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px

n,x_min,x_range,beta0,beta1,sigma = 20,5,5,2,3,5
x = stats.uniform(x_min, x_range).rvs(size=n)
errors = stats.norm(loc=0, scale=sigma).rvs(size=n)
Y = beta0 + beta1 * x + errors

df = pd.DataFrame({'x': x, 'Y': Y})
model_data_specification = smf.ols("Y~x", data=df) 
fitted_model = model_data_specification.fit() 

df['Data'] = 'Data' # hack to add data to legend 
fig = px.scatter(df, x='x',  y='Y', color='Data', 
                 trendline='ols', title='Y vs. x')

# This is what `trendline='ols'` is
fig.add_scatter(x=df['x'], y=fitted_model.fittedvalues,
                line=dict(color='blue'), name="trendline='ols'")

x_range = np.array([df['x'].min(), df['x'].max()])
y_line = beta0 + beta1 * x_range
fig.add_scatter(x=x_range, y=y_line, mode='lines',
                name=str(beta0)+' + '+str(beta1)+' * x', 
                line=dict(dash='dot', color='orange'))

# Add vertical lines for residuals
for i in range(len(df)):
    fig.add_scatter(x=[df['x'][i], df['x'][i]],
                    y=[fitted_model.fittedvalues[i], df['Y'][i]],
                    mode='lines',
                    line=dict(color='red', dash='dash'),
                    showlegend=False)
    
# Add horizontal line at y-bar
fig.add_scatter(x=x_range, y=[df['Y'].mean()]*2, mode='lines',
                line=dict(color='black', dash='dot'), name='y-bar')

fig.show(renderer="png") # USE `fig.show(renderer="png")` FOR ALL GitHub and MarkUs SUBMISSIONS
````

In the fitted model, the line chosen by the "ordinary least squares" (OLS) method is the one that best aligns with the observed data by minimizing the sum of the squared differences (residuals) between the actual Y values and the model’s predicted Y values. 

Squaring these residuals ensures that positive and negative differences don’t cancel each other out, and it also emphasizes larger errors, pushing the line to fit the data as closely as possible across all points. This approach yields the most accurate line for predicting Y based on x in the given dataset and model.

````python
# Let's define the data subset with long wait times as an example:
data_long_wait = old_faithful[old_faithful['waiting'] > long_wait_limit]

# Number of bootstrap samples
n_bootstraps = 160
bootstrap_slopes = []

# Bootstrap sampling
for _ in range(n_bootstraps):
    # Resample with replacement
    sample = data_long_wait.sample(frac=1, replace=True)
    
    # Fit linear regression model
    model = smf.ols('duration ~ waiting', data=sample).fit()
    
    # Collect the slope (coefficient of 'waiting')
    bootstrap_slopes.append(model.params['waiting'])

# Convert the slopes list to a NumPy array for easier handling
bootstrap_slopes = np.array(bootstrap_slopes)

# Plotting the bootstrap distribution of the slope coefficients using Plotly
fig = go.Figure(data=[go.Histogram(x=bootstrap_slopes, nbinsx=30, marker_color='blue', opacity=0.7)])
fig.update_layout(
    title="Bootstrap Sampling Distribution of Slope Coefficients",
    xaxis_title="Slope",
    yaxis_title="Frequency"
)
fig.show(renderer="png")
````

````python
# Create a copy of the long wait times subset for simulation
old_faithful_simulation = data_long_wait.copy()

# Simulate 'duration' values under the null hypothesis of no linear association
beta_0 = 1.65
beta_1 = 0
sigma = 0.37

# Generate simulated 'duration' values based on the null hypothesis
old_faithful_simulation['duration'] = beta_0 + beta_1 * old_faithful_simulation['waiting'] + \
                                      stats.norm(loc=0, scale=sigma).rvs(size=long_wait.sum())

# Fit a linear regression model to the simulated data and visualize the slope distribution
n_bootstraps = 160
simulated_slopes = []

for _ in range(n_bootstraps):
    # Bootstrap sample from the simulated data
    sample = old_faithful_simulation.sample(frac=1, replace=True)
    
    # Fit a linear regression model to the bootstrap sample
    model = smf.ols('duration ~ waiting', data=sample).fit()
    
    # Store the slope coefficient
    simulated_slopes.append(model.params['waiting'])

# Plotting the sampling distribution of the simulated slope coefficients using Plotly
fig = go.Figure(data=[go.Histogram(x=simulated_slopes, nbinsx=30, marker_color='orange', opacity=0.7)])
fig.update_layout(
    title="Sampling Distribution of Slope Coefficients Under Null Hypothesis (β₁ = 0) - Simulated Data",
    xaxis_title="Slope",
    yaxis_title="Frequency"
)
fig.show(renderer="png")
````

````python
# Calculate the 95% confidence interval for the simulated slopes
simulated_ci = np.percentile(simulated_slopes, [2.5, 97.5])

# Fit the original model on the long wait times subset (from actual data) to get the actual slope for comparison
original_model_long_wait = smf.ols('duration ~ waiting', data=old_faithful[long_wait]).fit()

# Extract the slope coefficient and p-value from the original model
original_slope = original_model_long_wait.params['waiting']
original_p_value = original_model_long_wait.pvalues['waiting']

# Check if the original slope is within the 95% confidence interval of the simulated slopes
is_slope_within_ci = simulated_ci[0] <= original_slope <= simulated_ci[1]

# Calculate the simulated p-value by comparing the original slope with the simulated slopes
simulated_p_value = np.mean([abs(slope) >= abs(original_slope) for slope in simulated_slopes])

# Display the results
print("95% Confidence Interval for the slope under null hypothesis:", simulated_ci)
print("Original slope:", original_slope)
print("Original p-value:", original_p_value)
print("Simulated p-value:", simulated_p_value)
print("Is the original slope within the 95% bootstrapped CI?", is_slope_within_ci)
````

> The first three assumptions associated with the **Simple Linear regression** model are that
> 
> - the $\epsilon_i$ **errors** (sometimes referred to as the **noise**) are **normally distributed**
> - the $\epsilon_i$ **errors** are **homoscedastic** (so their distributional variance $\sigma^2$ does not change as a function of $x_i$)
> - the linear form is [at least reasonably approximately] "true" (in the sense that the above two remain [at least reasonably approximately] "true") so that then behavior of the $Y_i$ **outcomes** are represented/determined on average by the **linear equation**)<br>
> 
>    and there are additional assumptions; but, a deeper reflection on these is "beyond the scope" of STA130; nonetheless, they are that<br><br>
> - the $x_i$ **predictor variable** is **measured without error**
> - and the $\epsilon_i$ **errors** are **statistically independent** (so their values do not depend on each other)
> - and the $\epsilon_i$ **errors** are **unbiased** relative to the **expected value** of **outcome** $E[Y_i|x_i]=\beta_0 + \beta_1x_i$ (which is equivalently stated by saying that the mean of the **error distribution** is $0$, or again equivalently, that the **expected value** of the **errors** $E[\epsilon_i] = 0$)
